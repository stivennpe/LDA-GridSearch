```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV
from pprint import pprint
```

```python
with open('Topic_00.txt', encoding="utf8") as f:
    src_content = [line.rstrip() for line in f]
```

```python
vectorizer = CountVectorizer(min_df= 100, max_df= 300, lowercase=bool, stop_words='english')
```

```python
df = vectorizer.fit_transform(src_content)
```

```python
lda = LatentDirichletAllocation(n_components = 10, random_state = 30, learning_method='online', learning_decay=0.9)
lda.fit(df)
```

```python
for index, topic in enumerate(lda.components_):
    print(f'Top 10 words for Topic #{index}')
    print([vectorizer.get_feature_names()[i] for i in topic.argsort()[-15:]])
    print('\n')

tes8 = pd.DataFrame(lda.components_)
for index, topic in enumerate(lda.components_):
    hola0 = (f'Topic #{index}')
    hola1=([vectorizer.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print(hola1)
tes8.index = hola1
tes8
```

```python
# Log Likelyhood: Higher the better
print("Log Likelihood: ", lda.score(df))
# Perplexity: Lower the better. Perplexity = exp(-1. * log-likelihood per word)
print("Perplexity: ", lda.perplexity(df))
# See model parameters
pprint(lda.get_params())
```

```python
# Define Search Param
search_params = {'n_components': [10, 15, 20, 25, 30], 'learning_decay': [.5, .7, .9]}
# Init the Model
lda = LatentDirichletAllocation(max_iter=5, learning_method='online', learning_offset=50.,random_state=0)
# Init Grid Search Class
model = GridSearchCV(lda, param_grid=search_params)
# Do the Grid Search
model.fit(df)
#GridSearchCV(cv=None, error_score='raise',
#       estimator=LatentDirichletAllocation(batch_size=128, doc_topic_prior=None,
#             evaluate_every=-1, learning_decay=0.7, learning_method=None,
#             learning_offset=10.0, max_doc_update_iter=100, max_iter=10,
 #            mean_change_tol=0.001, n_components=10, n_jobs=1,
  #           perp_tol=0.1, random_state=None,
   #          topic_word_prior=None, total_samples=1000000.0, verbose=0),
    #   iid=True, n_jobs=1,
     #  param_grid={'n_topics': [10, 15, 20, 25, 30], 'learning_decay': [0.5, 0.7, 0.9]},
      # pre_dispatch='2*n_jobs', refit=True, return_train_score='warn',
       #scoring=None, verbose=0)
```

```python
# Best Model
best_lda_model = model.best_estimator_
# Model Parameters
print("Best Model's Params: ", model.best_params_)
# Log Likelihood Score
print("Best Log Likelihood Score: ", model.best_score_)
# Perplexity
print("Model Perplexity: ", best_lda_model.perplexity(df))
```

```python
topicnames = ['Topic' + str(i) for i in range(best_lda_model.n_components)]
# Topic-Keyword Matrix
df_topic_keywords = pd.DataFrame(best_lda_model.components_)
# Assign Column and Index
df_topic_keywords.columns = vectorizer.get_feature_names()
df_topic_keywords.index = topicnames
# View
df_topic_keywords.head()
df_topic_keywords.transform(lambda s: s.nlargest(10).reindex(s.index, fill_value=0), axis='columns')
```
